```python
# Import all required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Problem Statement:
A retail company "ABC Private Limited" wants to understand the customer purchase behavior (specially, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products last months. The dataset also contains customer demographics (age, gender, marital status, city_type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.

Now, they want to build a model to predict the purchase amount of customer against products which will help them to create personalized offer for customers against different products.

```python
# Importing Datset
df_train = pd.read_csv('blackFriday_train.csv')
df_train.head()
```

```python
df_test = pd.read_csv('blackFriday_test.csv')
df_test.head()
```

```python
# Merge Both Train and Test dataset
df = df_train.append(df_test)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'], axis=1, inplace=True)
```

```python
df.head()
```

```python
pd.get_dummies(df['Gender'])
```

```python
pd.get_dummies(df['Gender'], drop_first=1)
```

```python
# Handling Categorical Feature - Gender
df['Gender'] = df['Gender'].map({'F':0, 'M':1})
```

```python
df.head()
```

```python
# Handling Categorical Feature - Age
df['Age'].unique()
```

```python
pd.get_dummies(df['Age'])
```

```python
pd.get_dummies(df['Age'], drop_first=True)
```

```python
df['Age'] = df['Age'].map({'0-17':1, '18-25':2, '26-35':3, '36-45':4, '46-50':5, '51-55':6, '55+':7})
df.head()
```

```python
# Fixing Categorical Feature - City_Category
pd.get_dummies(df['City_Category'])
```

```python
df_city = pd.get_dummies(df['City_Category'], drop_first=True)
df_city.head()
```

```python
df = pd.concat([df, df_city], axis=1)
df.head()
```

```python
# Drop Feature - City_Category
df.drop('City_Category', axis=1, inplace=True)
```
